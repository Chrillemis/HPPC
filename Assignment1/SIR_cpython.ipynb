```python
import SIR_python

S,I,R,beta, gamma, dt, tot_time = 999, 1, 0, 0.2, 0.1, 0.01, 100

my_sim = SIR_python.integrate_system(S,I,R,beta, gamma, dt, tot_time) 
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 2, figsize=(12, 10))

fig.text(0.5, 0.04, 'Time (days)', ha='center')
fig.text(0.04, 0.5, 'Number of individuals', va='center', rotation='vertical')

#Plot 1
data = np.genfromtxt('/home/lorentsen/HPPC/HPPC/hpc_course/module1/Unknown_SIR_data.csv', delimiter=',', skip_header = 1)

ax[0][0].plot(data[:, 0], data[:, 1], label='Susceptible (data)', color='blue')
ax[0][0].plot(data[:, 0], data[:, 2], label='Infected (data)', color='red')
ax[0][0].plot(data[:, 0], data[:, 3], label='Recovered (data)', color='green')
# ax[0][0].xlabel('Time')
# ax[0][0].ylabel('Number of individuals')
ax[0][0].legend()
# plt.show()
ax[0][0].title.set_text('Unknown SIR data')


#Plot 2
ax[0][1].plot(data[:, 0], np.sum(data[:, 1:], axis = 1), label = "Population in unkown data")
ax[0][1].plot(my_sim[:, 3], np.sum(my_sim[:, :3], axis = 1), label = "Population in my simulation")
# plt.legend()
# plt.xlabel('Time')
# plt.ylabel('Number of individuals')
ax[0][1].legend()
ax[0][1].title.set_text('Population in unknown data and simulation')

#Plot 3
S,I,R, dt, tot_time = 999, 1, 0, 0.01, 100
error = {"beta": [], "gamma": [], "MSE": []}
for beta in np.linspace(0.2, 0.3, 10):
    for gamma in np.linspace(0.05, 0.1, 10):
        my_sim = SIR_python.integrate_system(S,I,R,beta, gamma, dt, tot_time) 
        ax[1][0].plot(my_sim[:, 3], my_sim[:, 1])
        MSE = np.mean((my_sim[:, 1][::204] - data[:, 2])**2)
        error["beta"].append(beta)
        error["gamma"].append(gamma)
        error["MSE"].append(MSE)

error["MSE"] = np.array(error["MSE"])
error["beta"] = np.array(error["beta"])
error["gamma"] = np.array(error["gamma"])
ax[1][0].plot(data[:, 0], data[:, 2], label='Recovered (data)', color='black', linewidth=3, alpha = 0.5)

ax[1][0].title.set_text('Infected individuals for different beta and gamma values')

#Plot 4


my_sim = SIR_python.integrate_system(S,I,R, error["beta"][np.argmin(error['MSE'])], error["gamma"][np.argmin(error['MSE'])], dt, tot_time)
ax[1][1].plot(my_sim[:, 3], my_sim[:, 1], label='Infected (best fit)', color='orange', linewidth=3, alpha = 0.5)
ax[1][1].plot(my_sim[:, 3], my_sim[:, 0], label='Susceptible (best fit)', color='blue', linewidth=3, alpha = 0.5)
ax[1][1].plot(my_sim[:, 3], my_sim[:, 2], label='Recovered (best fit)', color='green', linewidth=3, alpha = 0.5)

ax[1][1].plot(data[:, 0], data[:, 1], label='Susceptible (data)', color='blue', linestyle='dashed')
ax[1][1].plot(data[:, 0], data[:, 2], label='Infected (data)', color='red', linestyle='dashed')
ax[1][1].plot(data[:, 0], data[:, 3], label='Recovered (data)', color='green', linestyle='dashed')

# ax[1][1].xlabel('Time')
# ax[1][1].ylabel('Number of individuals')
ax[1][1].legend()
ax[1][1].text(0, 400, f'Best fit parameters:\n beta = {error["beta"][np.argmin(error["MSE"])]:.3f}\n gamma = {error["gamma"][np.argmin(error["MSE"])]:.3f}\n MSE (I) = {min(error["MSE"]):.2f}')
ax[1][1].title.set_text('Best fit simulation vs data')
# plt.tight_layout()
```
